## 3. Data Cleaning

```python
train_data.head()
```

```python
home_ownership_mapping = {
    'RENT': 1,
    'MORTGAGE': 2,
    'OWN': 3,
    'OTHER': 4
}

loan_intent_mapping = {
    "EDUCATION": 1,
    "MEDICAL": 2,
    "PERSONAL": 3,
    "VENTURE": 4,
    "DEBTCONSOLIDATION": 5,
    "HOMEIMPROVEMENT": 6,
}

loan_grade_mapping ={
    "A": 1,
    "B": 2,
    "C": 3,
    "D": 4,
    "E": 5,
    "F": 6,
    "G": 7,
}

cb_person_default_on_file_mapping = {
    "N": 0,
    "Y": 1,
}
```

```python
def clean_data(input_data):
  input_data.drop("id",axis=1, inplace=True)
  input_data["person_home_ownership"] = input_data["person_home_ownership"].replace(home_ownership_mapping).astype(float)
  input_data["loan_intent"] = input_data["loan_intent"].replace(loan_intent_mapping).astype(float)
  input_data["loan_grade"] = input_data["loan_grade"].replace(loan_grade_mapping).astype(float)
  input_data["cb_person_default_on_file"] = input_data["cb_person_default_on_file"].replace(cb_person_default_on_file_mapping).astype(float)
  return input_data
```

```python
train_data = clean_data(train_data)
train_data
```

```python
test_data = clean_data(test_data)
test_data
```

## 4. Split the Data

```python
X = train_data.drop("loan_status", axis=1)
y = train_data["loan_status"]
```

```python
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## 6. Experimenting with different Algorithms!!!

### 1. Testing 1 with XGBoost and RandomForest!!!

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

XGB_v0 = XGBClassifier()
XGB_v0.fit(X_train, y_train)
```

```python
y_pred = XGB_v0.predict(X_test)
y_pred_prob = XGB_v0.predict_proba(X_test)[:, 1]
```

```python
accuracy_XGB_V0 = accuracy_score(y_test, y_pred)
roc_auc_XGB_V0 = roc_auc_score(y_test, y_pred_prob)
```

```python
print(f"Accuracy of XGBoost_v0 is: {accuracy_XGB_V0}")
print(f"Roc AUC score of XGBoost is: {roc_auc_XGB_V0}")
```

**Using RandomForest**

```python
RF_V0 = RandomForestClassifier()
RF_V0.fit(X_train, y_train)
```

```python
y_pred_RF_V0 = RF_V0.predict(X_test)
y_pred_prob_RF_V0 = RF_V0.predict_proba(X_test)[:, 1]
```

```python
y_pred_prob_RF_V0
```

```python
accuracy_RF_V0 = accuracy_score(y_test, y_pred_RF_V0)
roc_auc_RF_V0 = roc_auc_score(y_test, y_pred_prob_RF_V0)
```

```python
print(f"Accuracy of RF_V0 is: {accuracy_RF_V0}")
print(f"ROC AUC score of RF_V0 is: {roc_auc_RF_V0}")
```

### EValuating our Model

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
xgb_param_grid = {
    'n_estimators': [50, 100, 200, 500],
    'max_depth': [3, 4, 5, 6, 7],
    'learning_rate': [0.01, 0.1, 0.2, 0.3],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'gamma': [0, 0.1, 0.2, 0.3]
}


rf_param_grid = {
    'n_estimators': [50, 100, 200, 500],
    'max_depth': [None, 10, 20, 30, 40],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}
```

```python
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')

xgb_random_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=xgb_param_grid,
    n_iter=50,  # Number of random combinations to try
    scoring='roc_auc',  # Use AUC as evaluation metric
    cv=3,  # 3-fold cross-validation
    verbose=2,
    n_jobs=-1,  # Use all cores
    random_state=42
)
```

```python
xgb_random_search.fit(X_train, y_train)
print(f"Best XGBoost Params: {xgb_random_search.best_params_}")
```

```python
XGB_V1 = xgb_random_search.best_estimator_
y_pred_XGB_V1 = XGB_V1.predict(X_test)
y_pred_prob_XGB_V1 = XGB_V1.predict_proba(X_test)[:, 1]

print(f"XGBoost Accuracy: {accuracy_score(y_test, y_pred_XGB_V1)}")
print(f"XGBoost ROC AUC Score: {roc_auc_score(y_test, y_pred_prob_XGB_V1)}")
```

```python
rf_model = RandomForestClassifier(random_state=42)

rf_random_search = RandomizedSearchCV(
    estimator=rf_model,
    param_distributions=rf_param_grid,
    n_iter=50,  # Number of random combinations to try
    scoring='roc_auc',  # Use AUC as evaluation metric
    cv=3,  # 3-fold cross-validation
    verbose=2,
    n_jobs=-1,  # Use all cores
    random_state=42
)
```

```python
rf_random_search.fit(X_train, y_train)
print(f"Best Random Forest Params: {rf_random_search.best_params_}")
```

```python
RF_V1 = rf_random_search.best_estimator_
y_pred_RF_V1 = RF_V1.predict(X_test)
y_pred_prob_RF_V1 = RF_V1.predict_proba(X_test)[:, 1]
```

```python
print(f"Random Forest Accuracy: {accuracy_score(y_test, y_pred_RF_V1)}")
print(f"Random Forest ROC AUC Score: {roc_auc_score(y_test, y_pred_prob_RF_V1)}")
```
